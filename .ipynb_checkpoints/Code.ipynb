```c++17
int main(){
    unsigned long maximum = 0;
    unsigned long values[] = {10,20,30,40,50};
    for (unsigned long value: values)
    {
        if(value>maximum) maximum = value;
    }
    printf("The maximum value is %lu:\n", maximum);

    short array[] = {10, 02,30, 40, 560, 06, 04, 03, 50 ,330};
    size_t n_elements = sizeof(array)/sizeof(short); // old trick to print number of elements in an array in compile time. so there is no runtime cost.
    printf("The number of elements are %zu", n_elements); 
}
```
