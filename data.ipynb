```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("new.csv")
df["throughput"] = df["operations"]/(df["avg_time"])
```

```python
dfb = pd.read_csv("data.csv")
dfb["s_enq_throughput"] = dfb["s_enq"]/(dfb["avg_time"]-dfb["avg_timeout"])
dfb["s_deq_throughput"] = dfb["s_deq"]/(dfb["avg_time"]-dfb["avg_timeout"])
dfb["throughput"] = dfb["operations"]/(dfb["avg_time"]-dfb["avg_timeout"])

dfb["s_enq_throughput_norm"] = dfb["s_enq"]/(dfb["n_threads"]* (dfb["avg_time"]-dfb["avg_timeout"]))
dfb["s_deq_throughput_norm"] = dfb["s_deq"]/(dfb["n_threads"]* (dfb["avg_time"]-dfb["avg_timeout"]))
dfb["throughput_norm"] = dfb["operations"]/(dfb["n_threads"]* (dfb["avg_time"]-dfb["avg_timeout"]))
```

```python
fig, ax = plt.subplots(1,1)
# sns.lineplot(dfb, x = "n_threads", y = "throughput", hue = "name", ax = ax)
sns.lineplot(dfb, x = "n_threads", y = "s_enq_throughput_norm", hue = "name", ax = ax)
# sns.lineplot(dfb, x = "n_threads", y = "s_deq_throughput", hue = "name", ax = ax)
ax.set( yscale = "log")
```
